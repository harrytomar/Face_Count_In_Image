```python
import numpy as np
import math
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import cv2
import dlib

import scipy
%matplotlib inline
from scipy.spatial.distance import pdist
```

```python
root = "team7.jpg"
```

```python
img = cv2.imread(root)
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)#frame_HSV= cv2.cvtColor(frame,cv2.COLOR_RGB2HSV)  , cv2.COLOR_BGR2RGB
plt.imshow(img)
```

```python
%%time
li = []

arr = []
brr = []
ti = []
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")
plt.figure(figsize=(10,8))
faces = detector(img)
print(type(faces))

i=1

for face in faces:
    x1 = face.left()
    y1 = face.top()
    x2 = face.right()
    y2 = face.bottom()
    landmarks = predictor(img,face)
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(img," " + str(i),(x1,y1), font,2,(0,0,0),10,cv2.LINE_AA)
    i=i+1
    brr = []
    ti =[]
    for n in range(0,68):
        x = landmarks.part(n).x
        y = landmarks.part(n).y
        if n==60 or n==64:
            ti = [x,y]
            brr.append(ti)
        cv2.circle(img,(x,y),2,(255,0,0),1)
        print(x,y)
        li = [x,y]
        arr.append(li)
        cv2.rectangle(img,(x1,y1),(x2,y2),(0,255,0),4)
           
    
    
       
    
plt.imshow(img)
plt.show()
print(brr)
x6=brr[0][0]
y6=brr[0][1]
x7=brr[1][0]
y7=brr[1][1]
dist = math.sqrt((x7 - x6)**2 + (y7 - y6)**2)

print(dist)
print(arr)
crr=[]
crr=pdist(arr)
print(pdist(arr))
print(len(pdist(arr)))
print("number of faces are:-")
print(len(faces))
```
